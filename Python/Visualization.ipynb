```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

Line Plot: Create a line plot of the functions y = x and y = x² on the same graph. Add a legend to differentiate between the two lines.

```python
x = np.linspace(0, 50)
y1 = x
y2 = x**2

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = x')
plt.plot(x, y2, label='y = x²')
plt.legend()
plt.grid(True)
plt.show()
```

Scatter Plot: Create a scatter plot using random data for x and y. Color the points based on their value and add a color bar.

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

# Create plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar(scatter, label='Value')
plt.grid(True)
plt.show()
```

Subplots: Create a 2x2 grid of subplots. Each subplot should contain a different plot type (line plot, scatter plot, bar plot, and histogram) with some random data.

```python
x = np.linspace(0, 50)
y_line = np.sin(x)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
x_bar = np.arange(10)
y_bar = np.random.rand(10)
data_hist = np.random.randn(500)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
# Line plot
axs[0, 0].plot(x, y_line)
axs[0, 0].set_title('Line Plot')
# Scatter plot
axs[0, 1].scatter(x_scatter, y_scatter)
axs[0, 1].set_title('Scatter Plot')
# Bar plot
axs[1, 0].bar(x_bar, y_bar)
axs[1, 0].set_title('Bar Plot')
# Histogram
axs[1, 1].hist(data_hist, bins=30)
axs[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```

Bar Plot: Create a bar plot showing the average monthly temperatures of a city. Add error bars to represent the standard deviation of the temperatures.

```python
months = np.arange(1, 13)
temperatures = np.random.rand(12) * 30
errors = np.random.rand(12) * 5

plt.figure(figsize=(10, 6))
plt.bar(months, temperatures, yerr=errors, capsize=5, color='skyblue', edgecolor='black')
plt.title('Average Monthly Temperatures with Error Bars')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()

print("The numbers don't make sense because it is random")
```

Histogram: Create a histogram of 1000 data points drawn from a normal distribution. Add a KDE overlay to the histogram.

```python
data = np.random.randn(1000)

# Create plot
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='skyblue')
plt.grid(True)
plt.show()
```
