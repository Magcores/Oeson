```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\Escritorio_10_9_2022\Oeson\Python Oeson\Project 2 EDA\drowsiness_dataset.csv')
data
```

```python
#Create "fake" periods

#Define the number of periods
num_periods = 4

#Calculate the size of each period
period_size = len(data) // num_periods

# Create a period column
period_labels = ['Morning', 'Afternoon', 'Evening', 'Night']
data['period'] = pd.cut(data.index, bins=num_periods, labels=period_labels)

# Check the distribution of periods
print(data['period'].value_counts())

# Segment the data by period
morning_data = data[data['period'] == 'Morning']
afternoon_data = data[data['period'] == 'Afternoon']
evening_data = data[data['period'] == 'Evening']
night_data = data[data['period'] == 'Night']
```

```python
with pd.option_context('display.float_format', '{:.12f}'.format):
    print(data.describe())
```

```python
# Describe the data with floating-point format
with pd.option_context('display.float_format', '{:.12f}'.format):
    print(data.describe())
    print(morning_data.describe())
    print(afternoon_data.describe())
    print(evening_data.describe())
    print(night_data.describe())
```

```python
# Describe the data with floating-point format and print only the mean
with pd.option_context('display.float_format', '{:.12f}'.format):
    print("Overall Data Mean:")
    print(data.mean())

    print("\nMorning Data Mean:")
    print(morning_data.mean())

    print("\nAfternoon Data Mean:")
    print(afternoon_data.mean())

    print("\nEvening Data Mean:")
    print(evening_data.mean())

    print("\nNight Data Mean:")
    print(night_data.mean())
```

```python
# Check for missing values 
print(data.isnull().sum())
```

```python
#Histogram for numerical columns
data.hist(bins = 20, figsize=(12, 9))
plt.show()
```

```python
# Boxplot for numerical value
plt.figure(figsize=(15, 10))
sns.boxplot(data = data)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Separate numerical and categorical columns
numerical_cols = data.select_dtypes(include=['number']).columns
categorical_cols = data.select_dtypes(exclude=['number']).columns

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for numerical columns
Q1 = data[numerical_cols].quantile(0.25)
Q3 = data[numerical_cols].quantile(0.75)

# Calculate IQR for numerical columns
IQR = Q3 - Q1

# Determine the lower and upper bounds for filtering
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers in numerical columns
filtered_numerical_data = data[numerical_cols][~((data[numerical_cols] < lower_bound) | (data[numerical_cols] > upper_bound)).any(axis=1)]

# Combine filtered numerical data with original categorical data
filtered_data = pd.concat([filtered_numerical_data, data[categorical_cols]], axis=1).reindex(filtered_numerical_data.index)

# Display the filtered data
filtered_data
```

```python
with pd.option_context('display.float_format', '{:.12f}'.format):
    description = filtered_data.describe()
    print(description)
```

```python
# Correlation matrix
corr_matrix = filtered_data.corr()

# Heatmap of correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot bar plots for 'drowsiness' against other numerical variables
for col in numerical_cols:
    if col != 'drowsiness':
        plt.figure(figsize=(10, 6))
        sns.barplot(x=filtered_data['drowsiness'], y=filtered_data[col])
        plt.xlabel('Drowsiness')
        plt.ylabel(col)
        plt.title(f'Bar Plot: Drowsiness vs {col}')
        plt.show()
```

#### Analyzing and Viusalizing Correlations for Each Period: Calculate and visualize the correlation between drowsiness levels and other variables (heart rate and PPG readings) for each period separately

```python
def calculate_and_plot_correlations(data, period_name):
    correlation_heart_rate = data['drowsiness'].corr(data['heartRate'])
    correlation_ppg_green = data['drowsiness'].corr(data['ppgGreen'])
    correlation_ppg_red = data['drowsiness'].corr(data['ppgRed'])
    correlation_ppg_ir = data['drowsiness'].corr(data['ppgIR'])

    print(f'Correlation between drowsiness and heart rate ({period_name}): {correlation_heart_rate}')
    print(f'Correlation between drowsiness and PPG Green ({period_name}): {correlation_ppg_green}')
    print(f'Correlation between drowsiness and PPG Red ({period_name}): {correlation_ppg_red}')
    print(f'Correlation between drowsiness and PPG IR ({period_name}): {correlation_ppg_ir}')
    
    plt.scatter(data['heartRate'], data['drowsiness'], alpha=0.5, label='Heart Rate')
    plt.scatter(data['ppgGreen'], data['drowsiness'], alpha=0.5, label='PPG Green', color='green' )
    plt.scatter(data['ppgRed'], data['drowsiness'], alpha=0.5, label='PPG Red', color='red')
    plt.scatter(data['ppgIR'], data['drowsiness'], alpha=0.5, label='PPG IR', color='Purple')
    plt.title(f'Drowsiness levels vs Heart Rate and PPG Readings ({period_name})')
    plt.xlabel('Activity Level')
    plt.ylabel('Drowsiness level')
    plt.legend()
    plt.show()
    
# Calculate and plot correlations for each period
#calculate_and_plot_correlations(morning_data, 'Morning')
#calculate_and_plot_correlations(afternoon_data, 'Afternoon')
#calculate_and_plot_correlations(evening_data, 'Evening')
#calculate_and_plot_correlations(night_data, 'Night')
```
